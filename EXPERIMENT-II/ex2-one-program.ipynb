```python
import pandas as pd
import numpy as np
import sqlite3

# -------------------------------
# Create Sample Dataset
# -------------------------------
np.random.seed(42)

data = {
    "Student_ID": range(1, 101),
    "Name": [f"Student_{i}" for i in range(1, 101)],
    "Age": np.random.randint(18, 26, 100),
    "Gender": np.random.choice(["Male", "Female"], 100),
    "Department": np.random.choice(["CSE", "ECE", "IT", "MECH", "CIVIL"], 100),
    "Marks": np.random.randint(40, 101, 100),
    "Attendance": np.random.randint(60, 101, 100)
}

df = pd.DataFrame(data)

# Add Missing Values
df.loc[5, "Marks"] = np.nan
df.loc[20, "Age"] = np.nan
df.loc[35, "Department"] = np.nan
df.loc[50, "Attendance"] = np.nan

# -------------------------------
# Export Files
# -------------------------------
df.to_csv("students.csv", index=False)
df.to_excel("students.xlsx", index=False)
df.to_csv("students.txt", sep="\t", index=False)

print("Files Created Successfully!")

# -------------------------------
# Program 1 - Import CSV File
# -------------------------------
csv_df = pd.read_csv("students.csv")

print("\nCSV Dataset")
print(csv_df.head())

# -------------------------------
# Program 2 - First Five Records
# -------------------------------
print("\nFirst Five Records")
print(csv_df.head())

# -------------------------------
# Program 3 - Dataset Information
# -------------------------------
print("\nDataset Information")
print(csv_df.info())

# -------------------------------
# Program 4 - Statistical Summary
# -------------------------------
print("\nStatistical Summary")
print(csv_df.describe())

# -------------------------------
# Program 5 - Import Excel File
# -------------------------------
excel_df = pd.read_excel("students.xlsx")

print("\nExcel Dataset")
print(excel_df.head())

# -------------------------------
# Program 6 - Import SQL Data
# -------------------------------
conn = sqlite3.connect("students.db")

df.to_sql("StudentTable", conn, if_exists="replace", index=False)

sql_df = pd.read_sql("SELECT * FROM StudentTable", conn)

print("\nSQL Dataset")
print(sql_df.head())

conn.close()

# -------------------------------
# Program 7 - Import Data from Web
# -------------------------------
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"

web_df = pd.read_csv(url)

print("\nWeb URL Dataset")
print(web_df.head())

# -------------------------------
# Program 8 - Export to CSV
# -------------------------------
csv_df.to_csv("output.csv", index=False)
print("\nCSV File Exported")

# -------------------------------
# Program 9 - Export to Excel
# -------------------------------
csv_df.to_excel("output.xlsx", index=False)
print("Excel File Exported")

# -------------------------------
# Program 10 - Export to JSON
# -------------------------------
csv_df.to_json("output.json")
print("JSON File Exported")

# -------------------------------
# Additional EDA Operations
# -------------------------------
print("\nSeries")
print(csv_df["Marks"].head())

print("\nStudents with Marks > 80")
print(csv_df[csv_df["Marks"] > 80])

print("\nStudents from CSE Department")
print(csv_df[csv_df["Department"] == "CSE"])

print("\nStudents with Attendance > 90")
print(csv_df[csv_df["Attendance"] > 90])

print("\nStudents with Marks > 80 and Attendance > 90")
print(csv_df[(csv_df["Marks"] > 80) & (csv_df["Attendance"] > 90)])

print("\nSelected Columns")
print(csv_df[["Name", "Marks", "Department"]])

print("\nMissing Values Before Cleaning")
print(csv_df.isnull().sum())

csv_df["Age"] = csv_df["Age"].fillna(csv_df["Age"].mean())
csv_df["Marks"] = csv_df["Marks"].fillna(csv_df["Marks"].mean())
csv_df["Attendance"] = csv_df["Attendance"].fillna(csv_df["Attendance"].mean())
csv_df["Department"] = csv_df["Department"].fillna(csv_df["Department"].mode()[0])
csv_df["Gender"] = csv_df["Gender"].fillna(csv_df["Gender"].mode()[0])

print("\nMissing Values After Cleaning")
print(csv_df.isnull().sum())

csv_df.to_csv("students_cleaned.csv", index=False)

print("\nFinal cleaned dataset exported as students_cleaned.csv")
```
